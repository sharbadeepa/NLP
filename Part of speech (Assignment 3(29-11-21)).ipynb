```python
# 2nd Generate statistics for Brown corpus tagged data to answer the following questions:
# a.What proportion of word types are always assigned the same part-of-speech tag?

from nltk.corpus import brown
import nltk
brown_tagged_words = brown.tagged_words(categories='news', tagset='universal')
cfd = nltk.ConditionalFreqDist(brown_tagged_words)
conditions = cfd.conditions()
mono_tags = [condition for condition in conditions if len(cfd[condition]) == 1]
print(len(mono_tags))
print(len(conditions))
proportion_mono_tags = len(mono_tags) / len(conditions)
print(proportion_mono_tags)

# b.How many words are ambiguous, in the sense that they appear with at least two tags?
number_mono_tags = len(conditions) - len(mono_tags)
print(len(conditions))
print(number_mono_tags)

# c.What percentage of word tokens in the Brown Corpus involve these ambiguous
# words?
total_brown_words = set(brown.words())
number_shared_words = [word for word in mono_tags if word in total_brown_words]
percen_brown = len(number_shared_words) / len(brown.words())
print(percen_brown)

```

```python
# c. Identify three-word prepositional phrases of the form IN + DET + NN
brown_words = brown.tagged_words()
trigrams = list(nltk.trigrams(brown_words))
for trigram in trigrams:
    zipped_tag = [t for t in zip(*trigram)]
    if zipped_tag[1] == ('IN', 'DT', 'NN'):
        print(zipped_tag[0][:50])
```

```python
#5 Write a program to classify contexts involving the word “must” according to the tag of the
# following word. Can this be used to discriminate between the epistemic and deontic uses of
# “must”?

import nltk
from nltk.corpus import brown

tag_words = brown.tagged_words()
conditonal = nltk.ConditionalFreqDist(tag_words)
tag_sents = brown.tagged_sents()

bigrams = list(nltk.bigrams(tag_words))

for bigram in bigrams:
    zipped_tag = [list(t) for t in zip(*bigram)]
    if zipped_tag[0][0] == 'must' or zipped_tag[0][0] == 'Must':
        print(zipped_tag[0][0] + " " + zipped_tag[0][1])
        if zipped_tag[1][1] in ['BE', 'BE-HL', 'NN', 'NNS', 'NP-HL']:
            print("It is epistemic")
        elif zipped_tag[1][1] in ['HV', 'HV-TL', 'DO',  'RB', 'RB-HL', 'VB', 'VB-HL', 'VB-TL', 'VBZ']:
            print("It is deontic")
        else:
            print("It is unclear")



```

```python
# 6th question

import nltk
from nltk.corpus import brown

text = brown.words()
brown_taged_sents = brown.tagged_sents()
brown_tagged_words = brown.tagged_words()
fd = nltk.FreqDist(text)
cfd = nltk.ConditionalFreqDist(brown_tagged_words)
most_freq_words = fd.most_common(100)

likely_tags = dict((word, cfd[word].max()) for (word, _) in most_freq_words)
baseline_tagger = nltk.UnigramTagger(model=likely_tags)
print(baseline_tagger)

size = int(len(brown_taged_sents) * 0.9)
training_data = brown_taged_sents[:size]
test_data = (brown_taged_sents[size:])

uni_tagger = nltk.UnigramTagger(training_data)
print("uni_tagger",uni_tagger)
print(uni_tagger.evaluate(test_data))
```
