```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer
from nltk.stem import PorterStemmer
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import pandas as pd
stopwords = set(stopwords.words('english'))







file_lines = []
with open('/home/sharbadeepa/Downloads/reviews.txt', 'r') as f:
    lines=f.readlines()
    file_lines.append(lines)
    lines_lower = str(lines).lower()
    print(lines_lower)
    
    

word_tokenize = word_tokenize(lines_lower)
alpha = []
print(alpha)

snow_stemmer = SnowballStemmer(language='english')
snowball = []

Porter_Stemmer = PorterStemmer()
Porter = []

WordNet_Lemmatizer = WordNetLemmatizer()
lemmatizer = []



for word in word_tokenize:
    if word.isalpha():
        alpha.append(word)
        x = snow_stemmer.stem(word)
        snowball.append(x)
        y = Porter_Stemmer.stem(word)
        Porter.append(y)
        z = WordNet_Lemmatizer.lemmatize(word)
        lemmatizer.append(z)




        
print("snowball",snowball)
print("PorterStemmer", Porter)
print("lemmatizer",lemmatizer)





        
    
    
    
    
```

```python
import pandas as pd
import numpy as np


data = pd.DataFrame({'snowball': snowball,"stemming":Porter,"lemmatizer":lemmatizer}) 



print(data)







```

```python
out = []
for i,j,k in zip(data['snowball'],data['stemming'],data['lemmatizer']):
    if i == j and i == k and j == k:
        print('True')
        out.append('True')
    else:
        print('False')
        out.append('False')
```

```python
data['comparision'] = out
```
