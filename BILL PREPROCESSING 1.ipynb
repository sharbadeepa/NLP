```python
import cv2
import pytesseract
import os
import matplotlib.pyplot as plt

```

```python
img = cv2.imread('img1.png')

plt.imshow(img[:,:,::-1])
image_data = pytesseract.image_to_string(img)
print(image_data)
```

# Grey Scaling

```python
grey_scale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(grey_scale, cmap="gray")
image_data = pytesseract.image_to_string(grey_scale)
print(image_data)

```

# Thresholding

```python
bool_v, thrsh = cv2.threshold(grey_scale, 125, 255, cv2.THRESH_BINARY)
plt.imshow(thrsh, cmap="gray")
image_data = pytesseract.image_to_string(thrsh)
print(image_data)

```

# Thresholding OTSU

```python
ret, thresh2 = cv2.threshold(grey_scale, 150, 255, cv2.THRESH_OTSU) 
plt.imshow(thresh2, cmap="gray")
image_data = pytesseract.image_to_string(thresh2)
print(image_data)
```

#Contours

```python
imagePath =  "img1.png"
image = cv2.imread(imagePath)
imageCopy = image.copy()
blank_image = np.zeros((imageCopy.shape), dtype='uint8')
plt.imshow(blank_image)




```

```python
imageGray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(imageGray)
```

```python
img_bin = cv2.threshold(imageGray, 200, 255, cv2.THRESH_BINARY|cv2.THRESH_OTSU)[1]
plt.imshow(img_bin)

```

```python
contours, hierarchy = cv2.findContours(img_bin, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(blank_image, contours, -1, (255,255,255), 1)
plt.imshow(blank_image[:,:,::-1])
```

```python
contours, hierarchy = cv2.findContours(imageGray, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(image, contours, -1, (0,255,0), 3)
plt.imshow(image[:,:,::-1])
image_data = pytesseract.image_to_string(image)
print(image_data)

```
